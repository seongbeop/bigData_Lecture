```python
import pandas as pd

f = pd.read_csv("/content/wine.csv")

print(f.iloc[0,:])
```

```python
f.corr()
#상관계수를 구해주는 함수
#이는 각 컬럼끼리 페어하게 correiation coefficient값을 출력해준다
```

```python
X = f[['volatile acidity','residual sugar','chlorides','total sulfur dioxide','density','quality','class']]
y = f['alcohol']
#바로 위 셀에서 f.corr()으로 alcohol과 상관계수가 플러스마이너스 0.25 이상인 컬럼들만 확인하고
#f에서 'volatile acidity','residual sugar','chlorides','total sulfur dioxide','density','quality','class' 만 가져온다

X_train = X.iloc[:4000,:]
#모델 훈련에 쓸 데이터
X_valid = X.iloc[4000:,:]
#validation에 쓸 나머지 데이터
#X_train과 X_valid 의 범위를 지정 0~4000 , 4001~
y_train = y.iloc[:4000]
#훈련에 사용할 정답값 데이터 x와 매칭해야하므로 동일하게 4000개만 가져옴
y_valid = y.iloc[4000:]
#validation에 사용할 나머지 데이터
#y_train과 y_valid 의 범위를 지정 0~4000 , 4001~
```

```python
print(len(X_train))
print(len(X_valid))
print(len(y_train))
print(len(y_valid))
#데이터를 제대로 가져왔는지 갯수를 출력하여 확인
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
#PolynomialFeatures함수를 통해 데이터를 n의 제곱까지 피처를 늘릴준비를 한다 
#degree가 2이고, 데이터의 컬럼이 A,B 이렇게라면 A*2 ,B*2 , AB, A, B 이렇게 됨

poly_features.fit(X_train)
#Xtrain이 가지고 있는 컬럼이 몇개인지 파악, 몇개까지의 feature를 늘려야하는지 분석함(fit라는 말에 속으면 안된다)
poly_X= poly_features.transform(X_train)
#Xtrain은 원래 7개 이지만PolynomialFeatures떄문에 35개로 컬럼이 늘어난다 , 이를 poly_Xfksms tofhdns X로정의한다

#PolynomialFeatures함수를 통해 현재 데이터를 다항식 형태로 변경한다. (각 특성의 제곱 혹은 그 이상을 추가)
#degree 옵션으로 차수를 조절한다.
#include_bias 옵션은 True로 할 경우 0차항(1)도 함께 만든다.
#이렇게 만들어진 다항식 모델에 x를 fit transform하여 새로운 데이터를 생성한다
```

```python
from sklearn.linear_model import SGDRegressor
# sklearn에서 지원해주는 SGDRegressor 를 가져옴

sgd_reg = SGDRegressor(max_iter=50, tol=1e-3, penalty=None, eta0=0.1, random_state=42)
#SGDRegressor객체를 생성할건데 객체 안에 훈련 옵션을 셋팅한다, 
#경사하강법을 50번 반복 할 것이며(max_iter) , loss를 언제까지 줄일건지 (최대한 le-3 값 까지)
#이외의 옵션은 구글에서 검색해서 사람들이 대표적으로 사용하는 셋팅값을 불러와서 자주 쓴다
sgd_reg.fit(poly_X, y_train)

#poly_features를 적용한 35개 컬럼의 poly_X와 그에 대응하는 정답값인 y_train을 모델에 주고 SGD리그래서의 w들을 학습시킨다
```

```python
import numpy as np

X_new_poly = poly_features.transform(X_valid)
#PolynomialFeatures는 이미 X_train의 컬럼이 몇개고, 몇개까지 늘려야하는지 분석했다 
#굳이 X_valid로 또 분석 시킬 필요가 없다 (X_valid도 X_train이랑 피쳐 갯수가 똑같으니까)
#그래서 그냥 위에서 PolynomialFeatures 객체를 poly_feature로 만들걸 그대로 가져와서 fit 시키지 않고 transform만 x_valid한테 적용하면
#x_train과 똑같이 35개의 컬럼으로 늘어난다 . 그데이터를 X_new_poly라고 정의함 
```

```python
from sklearn.metrics import mean_squared_error
#MSE를 가지고 예측값과 실제값의 차이를 측정할 것이다
y_predict = sgd_reg.predict(X_new_poly)
#훈련이 완료된 sgd_reg에다가 valid set 이며 , PolynomialFeatures를 적용시킨 35개 컬럼의 x_new_poly를 넣고 prediction값을 뽑아낸다
print(mean_squared_error(y_predict,y_valid))
#MSE(예측값들 , 실제값들)
```
